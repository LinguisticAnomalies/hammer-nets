```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr

def make_auc_plot(train_con_auc, test_con_auc, train_diff_auc, test_diff_auc, train_ratio_auc, test_ratio_auc, layers):
    plt.rcParams["figure.figsize"] = (20,5)
    plt.plot(layers, train_con_auc, label="control train")
    plt.plot(layers, test_con_auc, label="control test")
    plt.plot(layers, train_diff_auc, label="c-d train")
    plt.plot(layers, test_diff_auc, label="c-d test")
    plt.plot(layers, train_ratio_auc, label="c/d train")
    plt.plot(layers, test_ratio_auc, label="c/d test")
    if max(layers) == 11:
        plt.xlabel("n-th layers")
        plt.ylabel("AUC")
        plt.title("AUC on Training and Test Set")
        plt.legend()
        plt.xticks(np.arange(0, 12, 1.0))
    else:
        plt.xlabel("first n layers")
        plt.ylabel("AUC")
        plt.title("AUC on Training and Test Set")
        plt.legend()
        plt.xticks(np.arange(1, 13, 1.0))
    for x, y in zip(layers, train_diff_auc):
        label = "{:.2f}".format(y)
        plt.annotate(label, # this is the text
                     (x,y), # this is the point to label
                     textcoords="offset points", # how to position the text
                     xytext=(0,10), # distance from text to points (x,y)
                     ha='center')
    for x, y in zip(layers, test_diff_auc):
        label = "{:.2f}".format(y)
        plt.annotate(label, # this is the text
                     (x,y), # this is the point to label
                     textcoords="offset points", # how to position the text
                     xytext=(0,10), # distance from text to points (x,y)
                     ha='center')
    for x, y in zip(layers, train_ratio_auc):
        label = "{:.2f}".format(y)
        plt.annotate(label, # this is the text
                     (x,y), # this is the point to label
                     textcoords="offset points", # how to position the text
                     xytext=(0,10), # distance from text to points (x,y)
                     ha='center')
    for x, y in zip(layers, test_ratio_auc):
        label = "{:.2f}".format(y)
        plt.annotate(label, # this is the text
                     (x,y), # this is the point to label
                     textcoords="offset points", # how to position the text
                     xytext=(0,10), # distance from text to points (x,y)
                     ha='center')
    plt.show()
    print("c-d model:")
    print("Pearson correction coeffificent: {}\np-value: {}".format(round(pearsonr(train_diff_auc, test_diff_auc)[0], 3),
                                                                    round(pearsonr(train_diff_auc, test_diff_auc)[1], 3)))
    print("c/d model:")
    print("Pearson correction coeffificent: {}\np-value: {}".format(round(pearsonr(train_ratio_auc, test_ratio_auc)[0], 3),
                                                                    round(pearsonr(train_ratio_auc, test_ratio_auc)[1], 3)))


    
def make_accu_plot(train_con_accu, test_con_accu, train_diff_accu, test_diff_accu, train_ratio_accu, test_ratio_accu, layers):
    plt.rcParams["figure.figsize"] = (20,5)
    plt.plot(layers, train_con_accu, label="control train")
    plt.plot(layers, test_con_accu, label="control test")
    plt.plot(layers, train_diff_accu, label="c-d train")
    plt.plot(layers, test_diff_accu, label="c-d test")
    plt.plot(layers, train_ratio_accu, label="c/d train")
    plt.plot(layers, test_ratio_accu, label="c/d test")
    if max(layers) == 12:
        plt.xlabel("n-th layers")
        plt.ylabel("Accuracy")
        plt.title("Accuracy on Training and Test Set")
        plt.legend()
        plt.xticks(np.arange(0, 12, 1.0))
    else:
        plt.xlabel("first n layers")
        plt.ylabel("Accuracy")
        plt.title("Accuracy on Training and Test Set")
        plt.legend()
        plt.xticks(np.arange(1, 13, 1.0))
    for x, y in zip(layers, train_diff_accu):
        label = "{:.2f}".format(y)
        plt.annotate(label, # this is the text
                     (x,y), # this is the point to label
                     textcoords="offset points", # how to position the text
                     xytext=(0,10), # distance from text to points (x,y)
                     ha='center')
    for x, y in zip(layers, test_diff_accu):
        label = "{:.2f}".format(y)
        plt.annotate(label, # this is the text
                     (x,y), # this is the point to label
                     textcoords="offset points", # how to position the text
                     xytext=(0,10), # distance from text to points (x,y)
                     ha='center')
    for x, y in zip(layers, train_ratio_accu):
        label = "{:.2f}".format(y)
        plt.annotate(label, # this is the text
                     (x,y), # this is the point to label
                     textcoords="offset points", # how to position the text
                     xytext=(0,10), # distance from text to points (x,y)
                     ha='center')
    for x, y in zip(layers, test_ratio_accu):
        label = "{:.2f}".format(y)
        plt.annotate(label, # this is the text
                     (x,y), # this is the point to label
                     textcoords="offset points", # how to position the text
                     xytext=(0,10), # distance from text to points (x,y)
                     ha='center')
    plt.show()
    print("c-d model:")
    print("Pearson correction coeffificent: {}\np-value: {}".format(round(pearsonr(train_diff_accu, test_diff_accu)[0], 3),
                                                                    round(pearsonr(train_diff_accu, test_diff_accu)[1], 3)))
    print("c/d model:")
    print("Pearson correction coeffificent: {}\np-value: {}".format(round(pearsonr(train_ratio_accu, test_ratio_accu)[0], 3),
                                                                    round(pearsonr(train_ratio_accu, test_ratio_accu)[1], 3)))
```
