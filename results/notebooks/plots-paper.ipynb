```python
import pickle
from pprint import pprint
import matplotlib.pyplot as plt
import numpy as np
```

```python
# functions for furthre useage
def mark_values(x_axis, y_axis):
    """
    mark the actul values on the plot
    :param x_axis: a list of x-axis values
    :type x_axis: list
    """
    for x, y in zip(x_axis, y_axis):
        label = "{:.2f}".format(y)
        plt.annotate(label, # this is the text
                     (x,y), # this is the point to label
                     textcoords="offset points", # how to position the text
                     xytext=(0,8), # distance from text to points (x,y)
                     ha='center')
```

```python
file_25 = "../evals/accumu_first_25_full.pkl"
plt.rcParams["figure.figsize"] = (20,10)
with open(file_25, "rb") as f:
    res_dict = pickle.load(f)
pprint(res_dict.keys())
plt.plot(list(range(1, 13)), res_dict["train_con_auc"],
         label="control model AUC on training set")
plt.plot(list(range(1, 13)), res_dict["test_con_auc"],
         label="control model AUC on test set")
plt.plot(list(range(1, 13)), res_dict["train_diff_auc"],
         label="c-d model AUC on training set")
plt.plot(list(range(1, 13)), res_dict["test_diff_auc"],
         label="c-d model AUC on test set")
plt.plot(list(range(1, 13)), res_dict["train_ratio_auc"],
         label="c/d model AUC on training set")
plt.plot(list(range(1, 13)), res_dict["test_ratio_auc"],
         label="c/d model AUC on test set")
mark_values(list(range(1, 13)), res_dict["train_con_auc"])
mark_values(list(range(1, 13)), res_dict["test_con_auc"])
mark_values(list(range(1, 13)), res_dict["train_diff_auc"])
mark_values(list(range(1, 13)), res_dict["test_diff_auc"])
mark_values(list(range(1, 13)), res_dict["train_ratio_auc"])
mark_values(list(range(1, 13)), res_dict["test_ratio_auc"])
plt.xticks(np.arange(1, 13, 1.0))
plt.title("Cumulative model with first zeroing 25% attention head in terms of AUC")
plt.legend()
#plt.show()
plt.savefig("accumu_first_25_auc.png", transparent=True)
```
